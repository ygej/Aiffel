# EXPLORATION_SBA : 2. Breast Cancer

```python
# 모듈 import
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

import pandas as pd
import numpy as np

# Data Load
breast_cancer = load_breast_cancer()
breast_cancer_data = breast_cancer.data
breast_cancer_label = breast_cancer.target

breast_cancer_df = pd.DataFrame(data = breast_cancer_data, columns = breast_cancer.feature_names)

print(breast_cancer_df.shape) # 총 569개의 데이터
breast_cancer_df.head() # columns = 30
```

```python
# Feature Names 확인
# 유방암 종양의 임상 데이터로, 종양의 특성이 각각의 columns이다.
print(breast_cancer.feature_names)
```

```python
print(breast_cancer.DESCR)
```

```python
# Target Names 확인
# malignant:악성, benign: 양성
print(breast_cancer.target_names)
```

```python
# Train, Test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data,
                                                   breast_cancer_label,
                                                   test_size = 0.4,
                                                   random_state = 31)
```

```python
#모델 학습

#DecisionTree
from sklearn.tree import DecisionTreeClassifier

d_tree = DecisionTreeClassifier(random_state = 31)

#RandomForest
from sklearn.ensemble import RandomForestClassifier

r_forest = RandomForestClassifier(random_state = 31)

# Support Vector Machine
from sklearn import svm

svm = svm.SVC()

# SGD CLassifier
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier()

# Logistic Regression
from sklearn.linear_model import LogisticRegression

logistic_r = LogisticRegression()

# 전체 모델 리스트
model_list = [d_tree, r_forest, svm, sgd, logistic_r]
```

```python
# 모델 학습 & 평가

# 모델 fit & accuracy score를 출력하는 class  

class model_fit_train:
    
    def __init__(self, var):
        self.fit = var.fit(X_train, y_train)
        self.pred = var.predict(X_test)
        
    def accuracy(self):
        print(accuracy_score(y_test, self.pred))
        
    def report(self):
        print(classification_report(y_test, self.pred))
        
    def confusion_matrix(self):
        print(confusion_matrix(y_test, self.pred))

# 모델별 accuracy score 출력        
        
for i in model_list:
    m = model_fit_train(i)
    m.accuracy()
```

```python
# 모델 선택
# accuracy score가 가장 높은 RandomForest 모델

m = model_fit_train(model_list[1])
m.report()
```

```python
# confusion_matrix
# 양성을 음성으로 판단하면 안 되기 때문에 Recall이 더 중요
# TP, FN, FP, TN
m.confusion_matrix()
```

```python
# 모델 선택
# accuracy score가 두번째로 높은 DecisionTree 모델

m = model_fit_train(model_list[0])
m.report()
```

```python
# confusion_matrix
# 양성을 음성으로 판단하면 안 되기 때문에 Recall이 더 중요
# TP, FN, FP, TN
m.confusion_matrix()
```

```python
# 결론: 정확도는 RandomForest 모델이 더 높지만, 본 데이터셋에서는 Recall이 더 높은 DecisionTree 모델 사용
```
